```python
from scipy.io.wavfile import write
import os

audio_dir = "/content/audio_files"
if not os.path.exists(audio_dir):
  os.makedirs(audio_dir, exist_ok=True)

for i, item in enumerate(dataset):
  audio_data = item['audio']['array']
  sampling_rate = item['audio']['sampling_rate']
  audio_path = item['audio']['path']

  target_audio_path = os.path.join(audio_dir, audio_path)

  # Lưu mảng âm thanh dứoi dạng tệp wav
  write(target_audio_path, sampling_rate, audio_data)

  print(f"Audio saved at: {target_audio_path}")

print(f"Đã lưu {i+1} tệp âm thanh vào {audio_dir}")
```

```python
from datasets import load_dataset

# Giả sử dataset hiện tại của bạn có tên là 'dataset'
# Chia trước 20% dữ liệu ra làm test + val (10% test + 10% val)
dataset_train_valtest = dataset.train_test_split(test_size=0.2, seed=42)

# Chia đôi phần test thành val và test, mỗi phần chiếm 10% tổng dữ liệu ban đầu
dataset_val_test = dataset_train_valtest['test'].train_test_split(test_size=0.5, seed=42)

# Lưu các splits lại
train_dataset = dataset_train_valtest['train']
val_dataset = dataset_val_test['train']
test_dataset = dataset_val_test['test']

# Kiểm tra số lượng từng split
print(f"Train size: {len(train_dataset)}")
print(f"Validation size: {len(val_dataset)}")
print(f"Test size: {len(test_dataset)}")
```

```python
import json
import librosa

def create_manifest(dataset, manifest_path, audio_dir):
  with open(manifest_path, 'w', encoding='utf-8') as f:
    for sample in dataset:
      audio_file = sample['audio']['path']
      audio_filepath = os.path.join(audio_dir, audio_file)
      sampling_rate = sample['audio']['sampling_rate']
      duration = librosa.get_duration(filename=audio_filepath, sr=sampling_rate)

      entry = {
          "audio_filepath": audio_filepath,
          "duration": duration,
          "text": sample['transcription']
      }

      f.write(json.dumps(entry, ensure_ascii=False) + '\n')
```

```python
manifest_dir = "/content/manifest"

if not os.path.exists(manifest_dir):
  os.makedirs(manifest_dir, exist_ok=True)

create_manifest(train_dataset, os.path.join(manifest_dir, "train.json"), audio_dir)
create_manifest(val_dataset, os.path.join(manifest_dir, "val.json"), audio_dir)
create_manifest(test_dataset, os.path.join(manifest_dir, "test.json"), audio_dir)
```

```python
import nemo.collections.asr as nemo_asr
from lightning.pytorch import Trainer
from omegaconf import OmegaConf, open_dict
import copy
import torch
```

```python
model = nemo_asr.models.ASRModel.from_pretrained("nvidia/stt_en_fastconformer_ctc_large")
```

```python
model.tokenizer
```

## Update vocabulary

```python
TOKENIZER_DIR = '/content/NeMo/scripts/tokenizers/output_vn/tokenizer_spe_unigram_v8000_pad_bos_eos'

model.change_vocabulary(new_tokenizer_dir=TOKENIZER_DIR, new_tokenizer_type="bpe")
```

```python
# Kiểm tra một dự đoán
text = "sự kết hợp hài hòa giữa đầm dạo phố và các trang phục khác mang lại sự thoải mái cho người mặc nhưng cũng không kém phần duyên dáng và nữ tính"
ids = model.tokenizer.text_to_ids(text)
decoded_text = model.tokenizer.ids_to_text(ids)

print(f"Tokenized text: {ids}")
print(f"Decoded text: {decoded_text}")
```

## Thiết lập cấu hình model

```python
# Cấu hình model
cfg = copy.deepcopy(model.cfg)
with open_dict(cfg):
  cfg.train_ds.manifest_filepath = '/content/manifest/train.json'
  cfg.validation_ds.manifest_filepath = '/content/manifest/val.json'
  cfg.test_ds.manifest_filepath = '/content/manifest/test.json'
  cfg.optim.lr = 0.0005
  cfg.train_ds.batch_size = 16
  cfg.validation_ds.batch_size = 16
  cfg.test_ds.batch_size = 16

# Thiết lập dữ liệu cho model
model.setup_training_data(cfg.train_ds)
model.setup_validation_data(cfg.validation_ds)
model.setup_test_data(cfg.test_ds)
model.cfg = cfg
```

```python
# Thiết lập Trainer và huấn luyện
import torch
import lightning.pytorch as ptl

if torch.cuda.is_available():
  accelerator = 'gpu'
else:
  accelerator = 'gpu'

EPOCHS = 50  # 100 epochs would provide better results

trainer = ptl.Trainer(devices=1,
                      accelerator=accelerator,
                      max_epochs=EPOCHS,
                      accumulate_grad_batches=1,
                      enable_checkpointing=False,
                      logger=False,
                      log_every_n_steps=5,
                      check_val_every_n_epoch=10)

# Setup model with the trainer
model.set_trainer(trainer)
```

# Training

```python
trainer.fit(model)

# 12. Lưu mô hình sau khi huấn luyện
model.save_to("resources/fastconformer-vietnamese.nemo")
```
